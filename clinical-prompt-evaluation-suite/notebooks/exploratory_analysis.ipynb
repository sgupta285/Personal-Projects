# Clinical Prompt Evaluation Suite\n\nThis notebook gives you a simple local workflow for inspecting demo runs after you seed data and execute an evaluation.

```
from app.db import SessionLocal\nfrom app import models\nfrom sqlalchemy import select
```

```
db = SessionLocal()\nruns = db.execute(select(models.EvaluationRun).order_by(models.EvaluationRun.id.desc())).scalars().all()\n[(run.id, run.status, run.aggregate_scores) for run in runs]
```

```
latest = runs[0]\nresults = db.execute(select(models.ItemEvaluationResult).where(models.ItemEvaluationResult.run_id == latest.id)).scalars().all()\n[(r.dataset_item_id, r.metric_scores, r.rubric_scores) for r in results]
```
