```python
# Import required libraries for data manipulation, plotting, and regression modeling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Set default figure size and plot style for consistency in all visualizations
plt.rcParams['figure.figsize'] = (10, 6)
sns.set_style('whitegrid')
```

```python
# Define base path to data directory
base_path = 'data'

# Load state-level datasets for happiness, insurance coverage, healthcare costs, and life expectancy
df_happy = pd.read_csv(f'{base_path}/happiest-states-2025.csv')
df_coverage = pd.read_csv(f'{base_path}/Health insurance coverage.csv')
df_costs = pd.read_csv(f'{base_path}/health-care-costs-by-state-2025.csv')
df_lifeexp = pd.read_csv(f'{base_path}/life-expectancy-by-state-2025.csv')
```

```python
# Display the first few rows of each dataset to confirm successful loading and inspect initial structure
print("\nHappiness Data:\n", df_happy.head(), "\n")
print("Coverage Data:\n", df_coverage.head(), "\n")
print("Healthcare Costs:\n", df_costs.head(), "\n")
print("Life Expectancy:\n", df_lifeexp.head(), "\n")
```

```python
# Standardize the column name for 'State' across all datasets to ensure compatibility for merging
for df in [df_costs, df_lifeexp, df_happy, df_coverage]:
    if 'Location' in df.columns:
        df.rename(columns={'Location': 'State'}, inplace=True)
    elif 'state' in df.columns:
        df.rename(columns={'state': 'State'}, inplace=True)

# Rename specific variables for clarity and brevity
df_costs.rename(columns={
    'HealthCareCostsSpendingPerCapita2020KFF': 'HealthcareCost'
}, inplace=True)

df_lifeexp.rename(columns={
    'LifeExpectancy_LifeExpectancyForBothSexes_yrs_2021': 'LifeExpectancy'
}, inplace=True)

df_happy.rename(columns={
    'HappiestStates_TotalHappinessScore_num_2024': 'HappinessScore'
}, inplace=True)

# Merge all datasets into a single DataFrame on 'State' using inner joins to retain common entries
df_merged = pd.merge(df_costs, df_lifeexp, on='State', how='inner')
df_merged = pd.merge(df_merged, df_happy, on='State', how='inner')
df_merged = pd.merge(df_merged, df_coverage, on='State', how='left')  # insurance optional

# Remove rows with missing essential data (healthcare cost or life expectancy)
df_merged.dropna(subset=['LifeExpectancy', 'HealthcareCost'], inplace=True)

# Display merged data shape and a preview of final dataset
print("Final merged DataFrame shape:", df_merged.shape)
print(df_merged.head())
```

```python
# Create side-by-side regression plots to visualize basic relationships:
# (1) Between healthcare spending and life expectancy
# (2) Between healthcare spending and happiness score

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot: Healthcare Cost vs Life Expectancy
sns.regplot(ax=axes[0], x='HealthcareCost', y='LifeExpectancy', data=df_merged,
            scatter_kws={'s':50}, line_kws={"color":"red"})
axes[0].set_title('Healthcare Cost vs Life Expectancy')
axes[0].set_xlabel('Healthcare Cost per Capita (USD)')
axes[0].set_ylabel('Life Expectancy (Years)')

# Plot: Healthcare Cost vs Happiness Score
sns.regplot(ax=axes[1], x='HealthcareCost', y='HappinessScore', data=df_merged,
            scatter_kws={'s':50}, line_kws={"color":"green"})
axes[1].set_title('Healthcare Cost vs Happiness Score')
axes[1].set_xlabel('Healthcare Cost per Capita (USD)')
axes[1].set_ylabel('Happiness Score')

# Adjust layout and display plots
plt.tight_layout()
plt.show()
```

```python
# Sort states by HappinessScore in descending order for clearer visualization
df_sorted_happy = df_merged.sort_values('HappinessScore', ascending=False)

# Create horizontal bar plot of happiness scores by state
plt.figure(figsize=(12, 8))
sns.barplot(x='HappinessScore', y='State', data=df_sorted_happy, orient='h')
plt.title("Happiness Score by State (2025)")
plt.xlabel("Happiness Score")
plt.ylabel("State")
plt.show()
```

```python
# OLS Regression: Simple model with Life Expectancy as a function of Healthcare Cost only
X_simple = df_merged[['HealthcareCost']]
y = df_merged['LifeExpectancy']
X_simple = sm.add_constant(X_simple)
model_simple = sm.OLS(y, X_simple).fit()

print("\n--- OLS Results: LifeExpectancy ~ HealthcareCost ---")
print(model_simple.summary())

# OLS Regression: Extended model controlling for HappinessScore
X_mult = df_merged[['HealthcareCost', 'HappinessScore']]
X_mult = sm.add_constant(X_mult)
model_mult = sm.OLS(y, X_mult).fit()

print("\n--- OLS Results: LifeExpectancy ~ HealthcareCost + HappinessScore ---")
print(model_mult.summary())
```

```python
# Define U.S. Census regions manually
northeast = ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut',
             'New York','New Jersey','Pennsylvania']
midwest   = ['Ohio','Michigan','Indiana','Wisconsin','Illinois','Minnesota','Iowa','Missouri',
             'North Dakota','South Dakota','Nebraska','Kansas']
south     = ['Delaware','Maryland','Virginia','West Virginia','Kentucky','North Carolina',
             'South Carolina','Georgia','Florida','Alabama','Mississippi','Tennessee','Arkansas',
             'Louisiana','Texas','Oklahoma']
west      = ['Montana','Idaho','Wyoming','Colorado','New Mexico','Arizona','Utah','Nevada',
             'California','Oregon','Washington','Alaska','Hawaii']

# Create a region label for each state
def get_region(state):
    if state in northeast: return 'Northeast'
    if state in midwest:   return 'Midwest'
    if state in south:     return 'South'
    if state in west:      return 'West'
    return 'Other'

df_merged['Region'] = df_merged['State'].apply(get_region)

# Create region dummy variables (South = baseline)
region_dummies = pd.get_dummies(df_merged['Region'], drop_first=True)

# Construct feature matrix with interaction terms: HealthcareCost  Region
X_region = pd.concat([
    df_merged[['HealthcareCost']].reset_index(drop=True),
    region_dummies.reset_index(drop=True)
], axis=1)

# Add interaction terms to allow the healthcare effect to vary by region
for col in region_dummies.columns:
    X_region[f'HealthcareCost_{col}'] = X_region['HealthcareCost'] * X_region[col]

# Ensure numerical formatting and drop any NA values
X_region = X_region.apply(pd.to_numeric, errors='coerce').astype(float)
X_region = sm.add_constant(X_region)

# Define dependent variable
y_region = pd.to_numeric(df_merged['LifeExpectancy'], errors='coerce').astype(float)

# Drop rows with NA in either X or y
valid_idx = X_region.dropna().index.intersection(y_region.dropna().index)
X_region = X_region.loc[valid_idx].astype(float)
y_region = y_region.loc[valid_idx].astype(float)

# Fit region-aware regression model with interaction terms
model_region = sm.OLS(y_region, X_region).fit()

print("\n--- OLS Results: LifeExpectancy ~ HealthcareCost + Region Dummies + Interactions ---")
print(model_region.summary())
```

```python
# Predict life expectancy values using the region-based model
y_pred = model_region.predict(X_region)
residuals = y_region - y_pred

# Store predictions and residuals in the original merged dataframe
plot_df = df_merged.copy()
plot_df['Predicted LifeExpectancy'] = y_pred
plot_df['Residuals'] = residuals

# Scatter plot: Actual vs Predicted values colored by Region
plt.figure(figsize=(10, 6))
sns.scatterplot(x=plot_df['LifeExpectancy'], y=plot_df['Predicted LifeExpectancy'],
                hue=plot_df['Region'], palette='tab10')
plt.plot([70, 85], [70, 85], 'k--')  # Line of perfect prediction
plt.title("Actual vs Predicted Life Expectancy by Region")
plt.xlabel("Actual Life Expectancy")
plt.ylabel("Predicted Life Expectancy")
plt.grid(True)
plt.legend(title='Region')
plt.tight_layout()
plt.show()

# Boxplot: Residuals by Region to examine model fit quality
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Residuals', data=plot_df)
plt.axhline(0, color='red', linestyle='--')  # Ideal residual line
plt.title("Residuals of Model by Region")
plt.ylabel("Prediction Error (Actual - Predicted)")
plt.xlabel("Region")
plt.tight_layout()
plt.show()
```

```python
# Visualize the relationship between Healthcare Cost and Life Expectancy, colored by Region
sns.lmplot(data=plot_df, x='HealthcareCost', y='LifeExpectancy', hue='Region',
           height=6, aspect=1.5, markers='o')
plt.title("Healthcare Cost vs Life Expectancy by Region")
plt.xlabel("Healthcare Cost per Capita")
plt.ylabel("Life Expectancy")
plt.tight_layout()
plt.show()
```
