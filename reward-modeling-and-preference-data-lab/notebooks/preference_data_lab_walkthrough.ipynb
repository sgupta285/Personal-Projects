# Preference Data Lab Walkthrough

This notebook shows a simple research loop: load the lab data, inspect example distributions, and compare annotator behavior.

```python
import sqlite3
import pandas as pd
from pathlib import Path

db_path = Path('preference_lab.db')
conn = sqlite3.connect(db_path)
examples = pd.read_sql_query('SELECT id, prompt_text, task_type, created_at FROM preference_examples', conn)
preferences = pd.read_sql_query('SELECT id, example_id, annotator_id, winner_candidate_id, created_at FROM preferences', conn)
examples.head(), preferences.head()
```

```python
preferences.groupby('annotator_id').size().rename('judgments').sort_values(ascending=False)
```
