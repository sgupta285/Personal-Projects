# Real-Time Control Analysis
Load the generated control log and visualize tracking and jitter.


```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(Path('..') / 'data/logs/run.csv')
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['time_s'], df['setpoint'], label='setpoint')
plt.plot(df['time_s'], df['true_speed'], label='true_speed')
plt.plot(df['time_s'], df['filtered_measurement'], label='filtered_measurement', alpha=0.8)
plt.legend()
plt.xlabel('Time (s)')
plt.ylabel('Speed')
plt.title('Control response')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(df['jitter_ms'], bins=40)
plt.title('Jitter histogram')
plt.xlabel('Jitter (ms)')
plt.ylabel('Loop count')
plt.show()
```
