# Dynamic Pricing Scenario Notebook
This notebook inspects the generated backtest outputs and recommendation queue.

```
from pathlib import Path
import json
import pandas as pd
import plotly.express as px
ROOT = Path.cwd().resolve().parents[0] if (Path.cwd()/'artifacts').exists() else Path('..')
artifacts = ROOT / 'artifacts'
recs = pd.read_csv(artifacts / 'top_recommendations.csv')
backtest = pd.read_csv(artifacts / 'backtest_rows.csv')
summary = json.loads((artifacts / 'backtest_summary.json').read_text())
summary
```

```
recs.head()
```

```
fig = px.histogram(recs, x='expected_profit_uplift_pct', nbins=30, title='Distribution of estimated profit uplift across recommendation queue')
fig.show()
```

```
category = backtest.groupby('category')['profit_uplift'].sum().reset_index().sort_values('profit_uplift', ascending=False)
px.bar(category, x='category', y='profit_uplift', title='Profit uplift by category').show()
```
