# Benchmark Analysis
This notebook loads generated scenario summaries and plots throughput, drop rate, and latency.


```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt
ROOT = Path('..').resolve()
DATA = ROOT / 'data' / 'generated'
summaries = []
for path in sorted(DATA.glob('*_summary.json')):
    if path.name.endswith('_replay_summary.json'):
        continue
    summaries.append(json.loads(path.read_text()))
summary = pd.DataFrame(summaries)
summary
```

```python
for path in sorted(DATA.glob('*_metrics.csv')):
    df = pd.read_csv(path)
    label = path.stem.replace('_metrics', '')
    plt.plot(df['seconds_since_start'], df['interval_throughput_msgs'], label=label)
plt.legend()
plt.title('Throughput by scenario')
plt.xlabel('seconds since start')
plt.ylabel('msgs/s')
plt.show()
```
